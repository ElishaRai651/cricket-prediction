```python
# Group by 'match_id' and 'innings' and sum only 'runs_off_bat' and 'extras'
total_score_df = delivery.groupby(['match_id', 'innings']).sum()[['runs_off_bat', 'extras']].reset_index()
```

```python
# Calculate the total score by summing 'runs_off_bat' and 'extras'
total_score_df['total_score'] = total_score_df['runs_off_bat'] + total_score_df['extras']
```

```python
# Add 1 run to the total score
total_score_df['total_score'] += 1
```

```python
#merging this total_score column based on match_id to match_info_data
match_df = match.merge(total_score_df[['match_id','innings','total_score']],left_on='id',right_on='match_id')
```

```python
teams = [
    'Mumbai Indians',
    'Chennai Super Kings',
    'Royal Challengers Bangalore',
    'Kings XI Punjab',
    'Kolkata Knight Riders',
    'Sunrisers Hyderabad',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['team2'] = match_df['team2'].str.replace('Gujarat Lions','Gujarat Titans')

match_df['team1'] = match_df['team1'].str.replace('Punjab Kings','Kings XI Punjab')
match_df['team2'] = match_df['team2'].str.replace('Punjab Kings','Kings XI Punjab')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
#extracting only match with no dl_applied
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','innings','city','winner','total_score']]
```

```python
delivery_df = match_df.merge(delivery, on=['match_id', 'innings'])
```

```python
delivery_df['current_score'] = delivery_df.groupby(['match_id', 'innings']).apply(
    lambda x: (x['runs_off_bat'] + x['extras']).cumsum()
).reset_index(level=[0, 1], drop=True)
```

```python
# Extract over and ball number from the ball column
delivery_df['over'] = delivery_df['ball'].astype(int)+1
delivery_df['ball'] = (delivery_df['ball'] * 10 % 10).astype(int)

# Cap the number of overs to 20
delivery_df = delivery_df[delivery_df['over'] < 20]
```

```python
# Filter for specific 'match_id' and 'innings'
filtered_df = delivery_df[(delivery_df['match_id'] == 1370352) & (delivery_df['innings'] == 2)]
filtered_df
```

```python
delivery_df.to_csv("new_data_with_over_score.csv", index=False)
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
#player_dismissed tell if batsman was out or not while playing that ball
#filling null values with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

#if column value is not zero mark it 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")

#changing datatype of column name
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

#finding wickets using cumulative sum function
wickets = delivery_df.groupby(['match_id', 'innings'])['player_dismissed'].cumsum()

#calculating wickets left by simple math operation
delivery_df['wickets_left'] = 10 - wickets
```

```python
# Filter for specific 'match_id' and 'innings'
filtered_df = delivery_df[(delivery_df['match_id'] == 1370352) & (delivery_df['innings'] == 2)]
filtered_df.tail()
```

```python
#crr=current_runs/current_overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
delivery_df.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df.head()
```

```python
groups = delivery_df.groupby(['match_id', 'innings'])

# Get unique combinations of match_id and innings
match_id_innings = delivery_df[['match_id', 'innings']].drop_duplicates().values.tolist()

last_five = []

for id in match_id_innings:
    match_group = groups.get_group(tuple(id))
    
    # Calculate the rolling sum of the current score over the last 30 balls
    last_five.extend(
        match_group['current_score'].rolling(window=30, min_periods=1).apply(lambda x: x.iloc[-1] - x.iloc[0], raw=False).tolist()
    )
delivery_df['last_five_over_score'] = last_five
```

```python
delivery_df.head()
```

```python
# Step 1: Calculate the wickets column
max_wickets = 10
delivery_df['wickets'] = max_wickets - delivery_df['wickets_left']

# Step 2: Calculate the last_five_over_wicket
groups = delivery_df.groupby(['match_id', 'innings'])

# Get unique combinations of match_id and innings
match_id_innings = delivery_df[['match_id', 'innings']].drop_duplicates().values.tolist()

last_five_over_wickets = []

for id in match_id_innings:
    match_group = groups.get_group(tuple(id))
    
    # Calculate the rolling sum of wickets over the last 30 balls
    last_five_over_wickets.extend(
        match_group['wickets'].rolling(window=30, min_periods=1).apply(lambda x: x.iloc[-1] - x.iloc[0], raw=False).tolist()
    )

# Ensure the lengths match before assigning
if len(last_five_over_wickets) == len(delivery_df):
    delivery_df['last_five_over_wicket'] = last_five_over_wickets
else:
    print("Length mismatch: Cannot assign 'last_five_over_wicket' column")

# Display the first few rows with the new columns
delivery_df.head()
```

```python
print(f'Before removing overs: {delivery_df.shape}')

# Group by match_id and innings
grouped = delivery_df.groupby(['match_id', 'innings'])

# Filter only groups where the maximum over is greater than 5
filtered_groups = grouped.filter(lambda x: x['over'].max() > 5)

# Keep only rows where the over is greater than 5
delivery_df = filtered_groups[filtered_groups['over'] > 5]

print(f'After removing overs: {delivery_df.shape}')
```

```python
X = final_df.drop(columns=['total_score'])
y = final_df['total_score']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
transformer = ColumnTransformer([
    ('transformer', OneHotEncoder(sparse_output=False, drop='first'),['batting_team','bowling_team', 'city'])
], remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', transformer),
    ('step2', StandardScaler()),
    ('step3', XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
